## 계산기 앱 만들어보기

```python
num1=int(input("정수를 입력하세요:"))
num2=int(input('정수를 입력해주세요:'))

print(num1*num2)
```

```python
num1=int(input("첫 번째 숫자: "))
op=input("연산자 (+, -, *, /): ")
num2=int(input("두 번째 숫자: "))

if op == '+':
    result = num1 + num2
elif op == '-':
    result = num1 - num2
elif op == '*':
    result = num1 * num2
elif op == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "0으로 나눌 수 없습니다."
else:
    result = "유효하지 않은 연산자입니다."
    
print("결과:", result)
```
